importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

```python
file_path = "/content/sample_data/Parkinsson disease.csv"
df = pd.read_csv(file_path)
print(df.head())
```

```python
#number of rows and columns in the dataframe
df.shape
```

```python
df.info()
```

```python
#getting statistical measures about the data
df.describe()
```

```python
#distribution of target variable
df['status'].value_counts()
```

0 --> healthy

1 --> parkinson

```python
#grouping the data based on the target variable
mean_by_status = df.groupby('status').mean(numeric_only=True)
print(mean_by_status)
```

Data preprocessing

Seperating the features and Target

```python
x = df.drop(columns=['name','status'],axis = 1)
y = df['status']
```

```python
print(x)
print(y)
```

Splitting the data into training data and splitting data


```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 2)
```

Data Standardization

```python
scalar = StandardScaler()
```

```python
scalar.fit(x_train)
```

```python
x_train = scalar.transform(x_train)
x_test = scalar.transform(x_test)
```

```python
print(x_train)
```

Model Training

Support Vector Machine Model

```python
model = svm.SVC(kernel = 'linear')
```

finding the hyperplane with x and y attributes

```python
#training the SVM model with training data
model.fit(x_train,y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy score on training data
x_train_pred = model.predict(x_train)
training_data_accuracy = accuracy_score(y_train,x_train_pred)
```

```python
print('Accuracy score of training data : ',training_data_accuracy)
```

```python
#accuracy score on testing data
x_test_pred = model.predict(x_test)
testing_data_accuracy = accuracy_score(y_test,x_test_pred)
```

```python
print('Accuracy score of testing data : ',testing_data_accuracy)
```

Building a Predictive System

```python
input_data = (197.07600,206.89600,192.05500,0.00289,0.00001,0.00166,0.00168,0.00498,0.01098,0.09700,0.00563,0.00680,0.00802,0.01689,0.00339,26.77500,0.422229,0.741367,-7.348300,0.177551,1.743867,0.085569)

input_data_as_numpy_array = np.asarray(input_data)

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

std_data = scalar.transform(input_data_reshaped)

prediction = model.predict(std_data)
print(prediction)


if (prediction[0] == 0):
  print("The Person does not have Parkinsons Disease")

else:
  print("The Person has Parkinsons Disease")
```
